```python
import pandas as pd
```

```python
p = r'Sample画像1_3件.pdf'
df = pd.DataFrame([p for i in range(5)], columns=['a'])
df['stem'], df['ext'] = df['a'].str.split('.').str[0], '.' + df['a'].str.split('.').str[1]
df['cumcount'] = (df.groupby('a').cumcount() + 1).astype(str).str.zfill(4)
df['a'] = df['stem'] + '_' + df['cumcount'] + df['ext']
df = df.drop(columns=['stem', 'ext', 'cumcount'])
df
```
